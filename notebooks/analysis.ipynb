```python
import pandas as pd
data = pd.read_json("../analysis/response_time.json")
```

```python
data
```

```python
attack_phase = data[data["phase"] == "attack"]
first_reinforcement_phase = data[data["phase"] == "first_reinforcement"]
reinforcement_phase = data[data["phase"] == "reinforcement"]
```

```python
print(first_reinforcement_phase["time"].mean())
print(reinforcement_phase["time"].mean())
print(attack_phase["time"].mean())
```

```python
reinforcement_phase['time'].mean() + attack_phase['time'].mean()
```

```python
reinforcements = data[data['phase'] == "reinforcement"]['time'].values
attacks = data[data["phase"] == "attack"]["time"].values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(attacks, 'o-')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(reinforcements, 'o-')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
